#**코로나 이후 ott 서비스 변화 분석**
- 코로나 전 후로 변경된 ott 서비스 이용 경험/ 빈도/시간 등의 변화 확인

- 이용 경험/사용빈도/사용시간 모두 다 코로나 이전인 2019년에 비해 증가했음을 확인해볼 수 있다.


```python
import pandas as pd
import numpy as np
import re
import urllib.request
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

###**OTT 서비스 이용 경험 변화**
- 2019년의 경우에는 이용 경험이 없는 이들(59%)이 있는 사람(41%)보다 많지만 2020 이후에는 경험이 있는 사람(72.2%)이 없는 사람(27.8%)보다 더 많은 비중을 차지한다.
- 코로나 이후 사람들의 ott 이용 경험이 높아졌다는 것을 알 수 있다.
- 2019-2020의 증가량(59%->72.2%)에는 미치지 못하지만,2020-2021의 증가량(72.2%->81.9%)또한 10퍼 이상으로 증가했다
  - 코로나 장기화 이후로 사용자들이 직접 방문하는 대신 ott 서비스를 사용하는 경우가 늘었다

```python
#데이터 다운로드
urllib.request.urlretrieve("https://raw.githubusercontent.com/world970511/elice_OTT/master/dataset/Covid_OTT/OTT%EC%84%9C%EB%B9%84%EC%8A%A4_%EC%9D%B4%EC%9A%A9_%EA%B2%BD%ED%97%98_20220106103142.csv",
                        filename="useOTT.txt")
use_data = pd.read_csv('useOTT.txt',encoding='utf-8')
use_data
```

```python
df=pd.DataFrame(columns=['year','사례수','이용경험','count'],
               data= [[2019,10864,'없음',int(10864*0.59)],[2019,10864,'있음',int(10864*0.41)],
                 [2020,10302,'없음',int(10302*0.278)],[2020,10302,'있음',int(10302*0.722)],
                 [2021,10154,'없음',int(10154*0.183)],[2021,10154,'있음',int(10154*0.817)]])

df.to_json("/content/drive/MyDrive/project/엘리스 OTT 분석/data/2019_2021이용경험.json", orient = 'index')
```

```python
d_2019=df.loc[df['year']==2019]
d_2020=df.loc[df['year']==2020]
d_2021=df.loc[df['year']==2021]

# 서브플롯
specs = [[{'type':'domain'}, {'type':'domain'},{'type':'domain'}]]
fig = make_subplots(rows=1, cols=3, specs=specs)

fig.add_trace(go.Pie(labels=d_2019['이용경험'], values=d_2019['count'], name="2019"),1, 1)
fig.add_trace(go.Pie(labels=d_2020['이용경험'], values=d_2020['count'], name="2020"),1, 2)
fig.add_trace(go.Pie(labels=d_2021['이용경험'], values=d_2021['count'], name="2021"),1, 3)

#도넛모양으로 만들기
fig.update_traces(hole=.4, hoverinfo="label+percent+name",textinfo='none')

fig.update_layout(
    title_text='2019-2021 사용경험 변화',
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='2019', x=0.127, y=0.5, font_size=20, showarrow=False),
                 dict(text='2020', x=0.5, y=0.5, font_size=20, showarrow=False),
                 dict(text='2021', x=0.87, y=0.5, font_size=20, showarrow=False),])
fig = go.Figure(fig)
fig.show()
```
